3.1 K-Means Clustering
Concepts:

Partitioning algorithm that minimizes the within-cluster sum-of-squares.
Sensitivity to initialization and the need for standardization.

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# -------------------------------
# Step 1: Simulate Banking Customer Data
# Assume features: Age and Annual Income (in USD)
np.random.seed(42)  # For reproducibility
n_samples = 300

# Simulate three distinct clusters representing different customer segments
# Cluster 1: Younger customers with lower income
cluster1 = np.random.normal(loc=[25, 30000], scale=[5, 5000], size=(100, 2))
# Cluster 2: Middle-aged customers with moderate income
cluster2 = np.random.normal(loc=[45, 60000], scale=[5, 7000], size=(100, 2))
# Cluster 3: Older customers with higher income
cluster3 = np.random.normal(loc=[65, 90000], scale=[5, 10000], size=(100, 2))

# Combine clusters into one dataset
data = np.vstack((cluster1, cluster2, cluster3))
df = pd.DataFrame(data, columns=['Age', 'Annual_Income'])

# -------------------------------
# Step 2: Data Preprocessing
# Standardize the data since clustering is sensitive to feature scales.
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)

# -------------------------------
# Step 3: Apply K-Means Clustering
# Here, we set n_clusters=3 based on our simulated data.
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(data_scaled)
df['KMeans_Cluster'] = clusters

# Evaluate the clustering using the silhouette score
score = silhouette_score(data_scaled, clusters)
print("K-Means Silhouette Score:", score)

# -------------------------------
# Step 4: Visualize the Clusters
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Annual_Income'], c=df['KMeans_Cluster'], cmap='viridis', alpha=0.7)
plt.title("Customer Segmentation with K-Means Clustering")
plt.xlabel("Age")
plt.ylabel("Annual Income (USD)")
plt.colorbar(label='Cluster Label')
plt.show()
```

Explanation:

We simulate a dataset with three clusters to mimic customer segmentation.
Data is standardized to ensure each feature contributes equally.
K-Means is applied, and clusters are evaluated with the silhouette score.
A scatter plot visualizes the clusters in the original feature space.
3.2 Hierarchical Clustering
Concepts:

Builds nested clusters by merging or splitting successively.
Produces a dendrogram to visualize the merging process.


```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# -------------------------------
# Step 1: Compute the Linkage Matrix
# Using the 'ward' method minimizes the variance within each cluster.
linked = linkage(data_scaled, method='ward')

# -------------------------------
# Step 2: Plot the Dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, truncate_mode='lastp', p=12)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Cluster Size")
plt.ylabel("Distance")
plt.show()

# -------------------------------
# Step 3: Apply Agglomerative Clustering
agg_clust = AgglomerativeClustering(n_clusters=3)
clusters_agg = agg_clust.fit_predict(data_scaled)
df['Hierarchical_Cluster'] = clusters_agg

# Optionally, visualize the hierarchical clustering results
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Annual_Income'], c=df['Hierarchical_Cluster'], cmap='plasma', alpha=0.7)
plt.title("Customer Segmentation with Hierarchical Clustering")
plt.xlabel("Age")
plt.ylabel("Annual Income (USD)")
plt.colorbar(label='Cluster Label')
plt.show()
```

Explanation:

The linkage function computes the hierarchical clustering.
A dendrogram is plotted to visually inspect how clusters merge.
Agglomerative clustering is then applied, and the resulting clusters are visualized.
3.3 DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
Concepts:

Groups together points that are closely packed and marks points that lie alone in low-density regions as outliers.
Parameters:
eps: Maximum distance between two samples to be considered as neighbors.
min_samples: Minimum number of samples in a neighborhood for a point to be considered as a core point.

```python
from sklearn.cluster import DBSCAN

# -------------------------------
# Step 1: Apply DBSCAN
# Note: Parameters 'eps' and 'min_samples' may need tuning based on the dataset.
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters_dbscan = dbscan.fit_predict(data_scaled)
df['DBSCAN_Cluster'] = clusters_dbscan

# -------------------------------
# Step 2: Visualize DBSCAN Clusters
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Annual_Income'], c=df['DBSCAN_Cluster'], cmap='coolwarm', alpha=0.7)
plt.title("Customer Segmentation with DBSCAN")
plt.xlabel("Age")
plt.ylabel("Annual Income (USD)")
plt.colorbar(label='Cluster Label')
plt.show()

# -------------------------------
# Note:
# DBSCAN may label some points as -1 indicating noise.
```

Explanation:

DBSCAN is particularly useful when the clusters have irregular shapes or when detecting outliers is important.
The visualization may reveal points marked as noise (typically with a cluster label of -1).
3.4 Gaussian Mixture Models (GMM)
Concepts:

A probabilistic model that assumes the data is generated from a mixture of several Gaussian distributions.
Provides soft clustering: each data point is assigned a probability of belonging to each cluster.

```python
from sklearn.mixture import GaussianMixture

# -------------------------------
# Step 1: Apply Gaussian Mixture Models
# We assume three components for the three customer segments.
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(data_scaled)
df['GMM_Cluster'] = gmm_labels

# -------------------------------
# Step 2: Visualize GMM Clusters
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Annual_Income'], c=df['GMM_Cluster'], cmap='coolwarm', alpha=0.7)
plt.title("Customer Segmentation with Gaussian Mixture Model")
plt.xlabel("Age")
plt.ylabel("Annual Income (USD)")
plt.colorbar(label='Cluster Label')
plt.show()
```

Explanation:

GMM offers a probabilistic view, which is valuable in finance where uncertainty and overlapping customer profiles are common.
Soft assignments provide insights into borderline cases between clusters.
4. Data Preprocessing for Clustering in Banking & Finance

Key Steps:

Handling Missing Values: Impute or remove missing data.
Feature Scaling: Standardization or normalization.
Outlier Detection: Remove or adjust extreme values that can skew clustering.

```python
# Assuming 'df' is your DataFrame with financial features
from sklearn.impute import SimpleImputer

# -------------------------------
# Step 1: Impute Missing Values (if any)
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

# -------------------------------
# Step 2: Standardize the Data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df_imputed)
```

Explanation:

This code ensures that missing data is handled and that all features are on a comparable scale—critical for distance-based algorithms like K-Means.
